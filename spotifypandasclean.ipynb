```python
a = dfbig.groupby(['uri'])['position'].min().rename('peakposition')  
b = dfbig.groupby(['uri'])['position'].count().rename('daysintop200')
c = dfbig[dfbig['position'] <= 10].groupby(['uri'])['position'].count().rename('daysintop10')
d = dfbig[dfbig['position'] <= 20].groupby(['uri'])['position'].count().rename('daysintop20')

dfextra = pd.concat([a, b,c,d],axis=1).fillna(0).reset_index()
dfextra.rename(columns={'uri':'Uri'},inplace=True)
dfextra.head()
```
